## Task 2 - Visualize the distributions of ratings and compensations

We'll now visualize the distribution of employee ratings and compensations to create the following plots:
     
     - A scatter plot of employees 

          age and employee salary,

          healthy eating and active lifestyle,

          healthy eating and salary

     - A count plot of groups
     
     - A histogram of employee salary

```python
#  Create a scatter plot of the 'age' and 'employee salary' features

plt.figure(figsize=(10, 5))

sns.scatterplot(data=data, x= 'age', y= 'salary', color= 'blue', edgecolor= 'linen', alpha= 0.5)

plt.title("Scatter plot of 'Employee Age' VS 'Employee Salary'")
plt.xlabel('Employee Age')
plt.ylabel('Employee Salary')
plt.show()
```

### Observations

- We observe that in general, there is a relationship in between 'age' and 'salary'.

- As age increases, the salary also increase, through there are some outliers present as well.

```python
# Create scatter plot of the 'healthy eating' and 'active lifestyle' features

plt.figure(figsize=(10, 5))

plt.scatter(data = data, x = 'healthy_eating', y = 'active_lifestyle', color='red', edgecolors= 'linen', alpha = 0.5)

plt.title("Scatterplot of 'Healthy Eating' VS 'Active Lifestyle'")
plt.xlabel('Healthy Eating')
plt.ylabel('Active Lifestyle')
plt.show()
```

### Observations

- We observe that in general, there is a relationship in between 'healthy eating' and 'active lifestyle'.

- As 'healthy eating' increases, the 'active lifestyle' also increases

```python
# Create scatter plot of the 'healthy eating' and 'salary' features

plt.figure(figsize=(10, 5))

plt.scatter(data = data, x = 'healthy_eating', y = 'salary', color='g', edgecolors= 'linen', alpha = 0.5)

plt.title("Scatterplot of 'Healthy Eating' VS 'Salary'")
plt.xlabel('Healthy Eating')
plt.ylabel('Salary')
plt.show()
```

### Observations

- We observe that in general, there is a relationship in between 'healthy eating' and 'salary' of employees.

- As 'healthy eating' increases, the 'salary' also increases.

```python
# Create a count plot of the 'groups' features

plt.figure(figsize=(10, 5))

sns.countplot(data = data, x= 'groups', edgecolor = 'linen', alpha= 0.7)

plt.title("Count Plot of Groups")
plt.xlabel('Groups')
plt.ylabel('Count')
plt.show()
```

### Observations

- Most of the employee is belong to Group 'A' or Group 'O', with Group 'A' having maximum freequency.

```python
# Create a histogram of the salary features

plt.figure(figsize=(11, 6))

sns.histplot(data = data, x = 'salary', color = 'teal', edgecolor = 'linen', bins= 10)

plt.title("Histogram of Salary")
plt.xlabel('Salary')
plt.ylabel('Count')
```

### Observations

- We see that salaries are not uniform, which is not necessarily a discrepency
- But we do observe that, there are employees with salaries on either extremities of the histogram.
- Based on histogram, we see that the count of highest earning employee is very less.

## Task 3 - Subset the Data based on Thresolds

We'll now the subsets of original data frames based on the following conditions:

     - Employees with healthy_eating > 8.
     - Employees with salary < 1000.
     - Employees with healthy_eating > 8 and salary < 1000.

```python
# Employees with healthy_eating > 8

subset1 = data[data['healthy_eating'] > 8]
subset1
```

```python
# Employees with salary < 1000

subset2 = data[data['salary'] < 1000]
subset2
```

```python
# Employees with healthy_eating > 8 and salary < 1000

subset3 = data[(data['healthy_eating'] > 8) & (data['salary'] < 1000)]
subset3
```

### Observations

1.   The only employee seemingly facing a discrepency in salary as compared to healthy_eating is employee who is having id - 26.

2.   The employee with id having 26 has a salary of 700.
