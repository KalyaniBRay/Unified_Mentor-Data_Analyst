## Task 2 - View the distributions of the various features in the data set and calculate their central tendencies

```python
# Create a histogram of the "Fixed Acidity" features

plt.figure(figsize= (11, 6))

sns.histplot(data= data, x= 'fixed acidity', color='orange', edgecolor = 'linen', alpha= 0.5, bins=5)
plt.title("Histogram of Fixed Acidity")
plt.xlabel('Fixed Acidity')
plt.ylabel('Count')
plt.show()
```

### Observations

     Histogram is normally distributed

     Maximum count of values for Fixed Acidity lies in between 6 to 8

Lrts see measure of Central Tendency :
     1. Mean
     2. Median
     3. Mode

```python
# Calculate the Mean of Fixed Acidity

round(data['fixed acidity'].mean(), 2)
```

```python
# Calculate the Median of Fixed Acidity

data['fixed acidity'].median()
```

```python
# Create a histogram of the "Fixed Acidity" features and also show the Mean and the Median

plt.figure(figsize=(11, 6))

sns.histplot(data=data, x='fixed acidity', color='orange', edgecolor='linen', alpha=0.5, bins=5)
plt.title("Histogram of Fixed Acidity")
plt.xlabel('Fixed Acidity')
plt.ylabel('Count')
plt.axvline(data['fixed acidity'].mean(), color='blue', linestyle='dashed', linewidth=2,label='Mean')
plt.axvline(data['fixed acidity'].median(), color='red', linestyle='dashed', linewidth=2,label='Median')
plt.legend()
plt.show()
```

```python
# Create a histogram of the "Volatile Acidity" features and also show the Mean and the Median

plt.figure(figsize= (11, 6))

sns.histplot(data= data, x= 'volatile acidity', color='g', edgecolor = 'linen', alpha= 0.5, bins=5)
plt.title("Histogram of Volatiley Acidity")
plt.xlabel('Volatile Acidity')
plt.ylabel('Count')
plt.show()
```

### Observations

this histogram is not well distributed, it is skewed a little towards the right

As we've seen skewness, therefore we can check the distribution using distplot function

```python
# Plot distplot using 'volatile acidity' feature

plt.figure(figsize=(11, 6))

sns.distplot(data['volatile acidity'], color='blue')

plt.title("Distplot of Volatile Acidity")
plt.xlabel('Volatile Acidity')
plt.ylabel('Density')
plt.show()
```

### Observations
The above plot shown the normal distribution

```python
# Calculate the skewness of 'volatile acidity'

data['volatile acidity'].skew()
```

### Observations

We can see that the skewness value is '>1', hence it is positively skewed

```python
#  Calculate the median 'volatile acidity' feature

data['volatile acidity'].median()
```

```python
# Create a histogram for showing 'volatile acidity' feature and also show the mean and the median

plt.figure(figsize= (11, 6))

sns.histplot(data= data, x= 'volatile acidity', color='g', edgecolor = 'linen', alpha= 0.5, bins=5)
plt.title("Histogram of Volatiley Acidity")
plt.xlabel('Volatile Acidity')
plt.ylabel('Count')
plt.axvline(data['volatile acidity'].mean(), color='blue', linestyle='dashed', linewidth=2,label='Mean')
plt.axvline(data['volatile acidity'].median(), color='red', linestyle='dashed', linewidth=2,label='Median')
plt.legend()
plt.show()
```

### Observations

The mean and the median are closed to each other and the different between them is very small

We can safely choose the mean as the central tendency here

```python
# Create a histogram for 'citric acid' feature

plt.figure(figsize= (11, 6))

sns.histplot(data= data, x= 'citric acid', color='r', edgecolor = 'linen', alpha= 0.5, bins=5)
plt.title("Histogram of citric acid")
plt.xlabel('citric acid')
plt.ylabel('Count')
plt.show()
```

```python
# Calculate the mean 'citric acid' features

data['citric acid'].mean()
```

```python
# Calculate the median 'citric acid' features

data['citric acid'].median()
```

```python
#  Calculate a histogram of the 'citric acid' features and also show the mean and median

plt.figure(figsize=(11, 6))

sns.histplot(data=data, x='citric acid', color='red', edgecolor='linen', alpha=0.5, bins=5)
plt.title("Histogram of Citric Acidity")
plt.xlabel('Citric Acidity')
plt.ylabel('Count')
plt.axvline(data['citric acid'].mean(), color='blue', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(data['citric acid'].median(), color='red', linestyle='dashed', linewidth=2, label='Median')
plt.legend()
plt.show()
```

### Observations

mean and median are close to each other and difference between them is very small

we can safely choose the mean and measure of the central tendency here

```python
# Calculate the distplot using 'citric acid' feature

plt.figure(figsize=(11, 6))

sns.distplot(data['citric acid'], color='blue')

plt.title("Distplot of Citric Acid")
plt.xlabel('Citric Acidity')
plt.ylabel('Density')
plt.show()
```

```python
# Create a count plot for the 'Quality' features

plt.figure(figsize=(11, 6))

sns.countplot(x='quality', data=data)

plt.title("Count Plot of Quality")
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

### Observations

It is quite clear that 6 is the highest count of quality whereas 9 is negligible

```python
#  Count the no of occurences of different catagories of the 'quality' features

data['quality'].value_counts()
```

```python
# Calculate the mode of the 'quality' features

data['quality'].value_counts().index[0]
```

## Task 3 - Create a new Pandas Series that contains the details of the acid types for a quality

```python
# Create a pandas series called 'rep_acid' that contains the details of the representative quality for the different types of acids

rep_acid = pd.Series(index= ['fixed acidity', 'volatile acidity', 'citric acid', 'quality'], data = [data['fixed acidity'].mean(), data['volatile acidity'].mean(), data['citric acid'].mean(), data['quality'].value_counts().index[0]])
```

```python
rep_acid
```

### <b>Observation from Task 3</b>

The Representative acid for the quality is as follows:

     - The mean value of the fixed acidity would be 6.854788
     - The mean value of the volatile acidity would be 0.278241
     - The mean value of the citric acidity would be 0.334192
     - The quality would be 6
